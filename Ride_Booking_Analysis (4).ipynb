**Average Time Comparison: Driver Arrival vs Trip Duration**

```python
avg_vtat = df['Avg VTAT'].mean()
avg_ctat = df['Avg CTAT'].mean()

print(f"Avg VTAT: {avg_vtat:.2f} minutes")
print(f"Avg CTAT: {avg_ctat:.2f} minutes")
```

```python
avg_df = pd.DataFrame({
    'Type': ['Driver to Pickup (VTAT)', 'Trip Duration (CTAT)'],
    'Avg Time (min)': [avg_vtat, avg_ctat]
})


# Plot
plt.figure(figsize=(6, 5))
sns.barplot(x='Type', y='Avg Time (min)', data=avg_df, palette='Blues')
plt.title("Average Time Comparison: VTAT vs CTAT")
plt.ylabel("Time (minutes)")
plt.xlabel("")
plt.xticks(rotation=10)
plt.tight_layout()
plt.show()
```

This chart compares two key time metrics:

VTAT (Vehicle to Arrival Time): The average time taken by the driver to reach the pickup location is approximately 8.46 minutes.

CTAT (Customer Trip Arrival Time): The actual trip duration from pickup to drop-off is significantly higher, averaging around 29.15 minutes.

The data clearly shows that while driver arrival is relatively quick, the majority of time is spent in the actual ride - an important insight for understanding operational efficiency and customer time investment.

**Cancellation by the customer**

```python
cancellation_counts = df['Reason for cancelling by Customer'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(
    cancellation_counts,
    labels=cancellation_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    shadow=True
)


plt.title('Reason for cancelling by Customer - Pie Chart')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Vehicle Type: 'Auto' is the most popular vehicle type, followed by 'Go Mini' and 'Go Sedan'. This indicates a strong preference for economical and compact options among users.

Payment Method: UPI is overwhelmingly the most preferred payment method, highlighting the importance of seamless digital payment options.

Pickup Locations: You've identified the top and least popular pickup locations, which can be valuable for optimizing driver distribution and identifying areas for potential growth or targeted marketing.

Average Time Comparison: The average trip duration (CTAT) is significantly longer than the average time it takes for a driver to reach the pickup location (VTAT). This is expected but emphasizes that the majority of the customer's time is spent in the ride itself.

Customer Cancellations: The pie chart shows the distribution of reasons for customer cancellations. "Wrong Address" is the most frequent reason, which could indicate issues with location accuracy or user input during booking.
